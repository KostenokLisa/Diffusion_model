```python
import numpy as np
from numpy.random import multivariate_normal
from numpy.random import binomial
from math import sqrt
```

```python
M = 10000
mean1 = [0.0]
mean2 = [-10.0]
var1 = [10.0]
var2 = [2.0]
p = 0.1
```

```python
M = 500
mean1 = [0.0, 1.0]
mean2 = [-10.0, -9.0]
var1 = [10.0, 10.0]
var2 = [2.0, 4.0]
p = 0.1
```

```python
idx1 = binomial(1, p, M)
idx2 = np.ones(M) - idx1

data1 = multivariate_normal(mean1, np.diag(np.sqrt(var1)), M)
data2 = multivariate_normal(mean2, np.diag(np.sqrt(var2)), M)
#print(data1)

gmm_data = np.diag(idx1) @ data1 + np.diag(idx2) @ data2
    
print(gmm_data)
```

Для одномерного распределения с p = 0.1

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(30,20))
N = len(gmm_data)

ax.hist(gmm_data, N // 50)
ax.hist(data1, N // 50, alpha = 0.5)
ax.hist(data2, N // 50, alpha = 0.5)
plt.title("1dim distribution")
```

Для двумерного распределения с p = 0.1

```python
plt.scatter(gmm_data.T[0], gmm_data.T[1])
plt.title("2dim distribution")
```

```python
dim = gmm_data.shape
N = len(dim) - 1
if N == 1:
    step = 100
    fig, ax = plt.subplots(figsize=(30, 20))
    ax.hist(gmm_data, step)
```
